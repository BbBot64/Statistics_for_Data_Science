```python
def split_train_test(data: pd.DataFrame, test_ratio: float) -> pd.DataFrame:
    """Define a function to create a training and test set
    Input: Takes Pandas DataFrame and test split ratio (eg. )
    Output: Train and Test datasets (DataFrames)"""
    np.random.seed(42) # change seed if you want to generate different results
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_dataset, test_dataset = split_train_test(data=original_spam_dataframe, test_ratio=0.2)
```

## Classifiers
Classifiers being demonstrated include:
* Decision Tree Classifier as a benchmark

Metrics/Tools being used to quantify each classifier include:
* Accuracy Score

```python
# Classifiers
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

# Metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

# Version
sklearn.__version__
```

```python
def rate_classifier(test_target_data, test_target_predict):
    print("Confusion Matrix\n", confusion_matrix(test_target_data, test_target_predict))
    print("Classification Report\n", classification_report(test_target_data, test_target_predict))
    print("Accuracy Score\n", accuracy_score(test_target_data, test_target_predict))
```
